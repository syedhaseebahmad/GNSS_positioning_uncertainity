```python
#Reading the training data with OLS method 
all_data = pd.read_csv('C:/Users/Haseeb/Desktop/internship/Data/5_cities.csv')
#test_data=pd.read_csv('TestingData_WLS_variance.csv')
```

```python
all_data.head()
```

```python
data = all_data.iloc[:1000,[24,8,14,16,67,49,79,80,84,85,86,90,97,102,103,109,110,111,89]]
```

```python
data.head()
```

```python
new_column_names =  [ 'Psr','Crp', 'Doppler', 'Snr', 'Elevation', 'DCrp', 'WlsPsrR','WlsNPsrR', 'D_Psr_Crp', 'D_Doppler_DCrp', 'PRC', 'LT_CN0', 'Sigma_CN0','nbsat', 'SFM', 'variance_snr', 'variance_PrCorrected', 'variance_PrR','Uncertainty with WLSElevation']
    
data.columns = new_column_names
```

```python

#Dropping Nan Values From the feature Pseudorange residuals 
#(The missing data comes from the epoques where the number of satellite less than 4 )
# to calculate  The pseudorange residuals needs at least 4 satellites   
data.dropna(subset = ["WlsNPsrR"], inplace=True)
#test_data.dropna(subset = ["WlsNPsrR"], inplace=True)
```

```python
target  =  data['Uncertainty with WLSElevation']
```

```python
data.drop(columns=['Uncertainty with WLSElevation'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data,target, test_size = 0.20, random_state =42)
    
print("Shape of X_train: ", X_train.shape )
print("Shape of X_test: ", X_test.shape )
print("Shape of y_train: ",y_train.shape )
print("Shape of y_test: ", y_test.shape)
```

```python
forest_model = RandomForestRegressor(n_estimators=100, criterion='squared_error', max_depth=110, min_samples_split=2, min_samples_leaf=1,
                                        min_weight_fraction_leaf=0.0, max_features='sqrt', max_leaf_nodes=None, min_impurity_decrease=0.0, bootstrap=False, 
                                       oob_score=False, n_jobs=None, random_state=None, verbose=0, warm_start=False, ccp_alpha=0.0, max_samples=None)
```

```python
forest_model.fit(X_train, y_train)
```

```python
pickled_model = pickle.load(open('randomforestmodel.pkl','rb'))
```
